# 变为单目标

```python
import Bgolearn.BGOsampling as BGOS
import pandas as pd


data = pd.read_csv('TrainingData.csv')

x = data.iloc[:,:-2]
y = data.iloc[:,-2] * data.iloc[:,-1]

dd = pd.read_csv('DesignData.csv') 

Bgolearn = BGOS.Bgolearn()

Mymodel = Bgolearn.fit(data_matrix = x, Measured_response = y, virtual_samples = dd)

Mymodel.EI()
```

# 帕眉托前沿

## 1: 仅使用property1 做推荐

```python
import Bgolearn.BGOsampling as BGOS
import pandas as pd


data = pd.read_csv('TrainingData.csv')

x = data.iloc[:,:-2]
y = data.iloc[:,-2]

dd = pd.read_csv('DesignData.csv') 

Bgolearn = BGOS.Bgolearn()

Mymodel = Bgolearn.fit(data_matrix = x, Measured_response = y, virtual_samples = dd)

score_1, _= Mymodel.EI_plugin()
```

```python
score_1
```

## 2: 仅使用property2 做推荐

```python
import Bgolearn.BGOsampling as BGOS
import pandas as pd


data = pd.read_csv('TrainingData.csv')

x = data.iloc[:,:-2]
y = data.iloc[:,-1]

dd = pd.read_csv('DesignData.csv') 

Bgolearn = BGOS.Bgolearn()

Mymodel = Bgolearn.fit(data_matrix = x, Measured_response = y, virtual_samples = dd)

score_2, _= Mymodel.EI_plugin()
```

```python
score_2
```

## 画出前沿

```python
import matplotlib.pyplot as plt
plt.xlabel('Score 1')
plt.ylabel('Score 2')
plt.scatter(score_1,score_2)
```
